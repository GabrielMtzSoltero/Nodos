Import necessary packages

```python
import numpy as np
import random as rd
import scipy.io
import os
```

Definition of function that generates de matrix

```python

def generateConectionMatrix(size=20,probability=[0,1],filename=""):
    #creeate a matriz of zeros
    matrix=np.zeros([size,size])
    #initialize the values for rows and columns just to make clear the code
    rows=columns=size
    for row in range(rows):
        #start the for in the row+1 values to not take the position under the diagonal
        for column in range(row+1,columns):
            #choose between o or 1 with their respective probability
            conexion=np.random.choice([0,1], 1, p=probability)[0]
            matrix[row][column]=conexion
            matrix[column][row]=conexion

    #get the sum o the ones on each row (axis 1 is the one for the columns)
    conectionsPerNode=np.sum(matrix,axis=1)
    for row in range(rows):
        matrix[row][row]=conectionsPerNode[row]*-1
    eigenValues=np.linalg.eigh(matrix)[0]
    if filename!="":
        print(os.path.basename(filename).rsplit('.', 1)[1])
        if os.path.basename(filename).rsplit('.', 1)[1]=="mat":
            data={'A':matrix}
            scipy.io.savemat(filename,data)
        else:
            np.save(filename,matrix)
    return matrix,eigenValues
```





*   First parameter default value is 20 for size
*   Second parameter for probabilities is [0,1] wich means all nodes are conected
*   Third parameter is the file name in case you need to save the generated matrix, by default its gonna save it as npy file but you can also save it as a matlab file. The file can be dowloaded from the files tab
